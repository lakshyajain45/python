```python
#1 Staircase Counter 🪜
def factorial(n):
    if n==0 or n==1:
      return 1
    else:
      return n*factorial(n-1)

def count_ways(n):
    if n==0:
      return 1
    elif n==1:
      return 1
    else:
      return count_ways(n-1)+count_ways(n-2)

print(factorial(5))
print(factorial(0))
print(count_ways(1))
print(count_ways(3))
print(count_ways(5))
```

```python
#2 Movie Leaderboard 🎬
movies = [
    {"title": "Interstellar",    "year": 2014, "rating": 8.7},
    {"title": "Inception",       "year": 2010, "rating": 8.8},
    {"title": "The Dark Knight", "year": 2008, "rating": 9.0},
    {"title": "Parasite",        "year": 2019, "rating": 8.5},
    {"title": "Dune",            "year": 2021, "rating": 8.0},
]
a=[]
b=[]
c=[]
d=[]
e=[]
f=[]
for i in movies:
    a.append(i['rating'])
    c.append(i['year'])
    e.append(i['title'])
c.sort()
e.sort()
a.sort(reverse=True)
for i in range(len(a)):
    for j in movies:
        if a[i]==j['rating']:
            b.extend([j['title'],j['rating']])
        if c[i]==j['year']:
            d.extend([j['title'],j['year']])
for k in movies:
    if k['rating']>8.5:
        f.append(k['title'])

print(b)
print(d)
print(e)
print(f)
```

```python
#4 Build Your Own Sorter 🔃
def bubble_sort(arr):
    n = len(arr)
    c=0
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
                c=c+1
    print('num of swap',c)
    return arr

nums   = [64, 34, 25, 12, 22, 11, 90]
scores = [45, 92, 38, 71, 85]
ex=[1,2,3,4]
print(bubble_sort(nums))
print(bubble_sort(scores))
print(bubble_sort(ex))
```
