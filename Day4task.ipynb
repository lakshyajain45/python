```python
#Q1 Username Validator 🔐
def validate_username(username):
  if len(username)>=5 and len(username)<=15:
    if username[0].isalpha():
      if username.isalnum() or username.isalpha() or '_' in username:
        return "Valid username"
      else:
        return "only letters,digits,underscore allowed"
    else:
      return "Must start with a letter"
  elif len(username)<5:
      return "Too Short"
  else:
      return "Too Long"
print(validate_username("rahul_99"))
print(validate_username("hi"))
print(validate_username("99abc"))
print(validate_username("hello world"))
print(validate_username("this_username_is_way_too_long_bro"))
```

```python
#Q2 Word Frequency Counter 📊
s = "the cat sat on the mat the cat sat"
s=s.split(' ')
d={}
c={}
m=1
e=set()

for i in s:
  d.update({i:s.count(i)})
  if s.count(i)>m:
    m=s.count(i)
    c.update({i:m})
  if s.count(i)>1:
    e.add(i)
print('words',s)
print('frequency',d)
print('most frequent',c)
print('Repeated',e)


```

```python
#Q3 Number Pattern Printer 🔢
def print_triangle(n):
    for i in range(1,n+1):
        for j in range(1,i+1):
            print("*",end="")
        print()
def print_square(n):
    for i in range(n):
        for j in range(n):
            print("*",end="")
        print()

def print_reverse(n):
    for i in range(1,n+1):
        for j in range(1,7-i):
            print("*",end="")
        print()

print_triangle(5)
print()
print_square(4)
print()
print_reverse(5)
```

```python
#Q5 Safe ATM Machine 🏧
def withdraw(balance, amount):
  try:
     if amount <= 0:
      print("Invalid amount!")
     elif amount > balance:
      print("Insufficient balance!")
     else:
      balance = balance - amount
      print('Withdrew ₹',balance, 'Remaining balance: ₹',balance)

     return balance
  except Exception:
        print("Transaction error!")
        return balance

def atm_session():
    balance = 5000
    print(f"Welcome! Your balance: ₹{balance}")

    balance = withdraw(balance, 1000)
    balance = withdraw(balance, 9999)
    balance = withdraw(balance, -500)
    balance = withdraw(balance, 0)

atm_session()
```
