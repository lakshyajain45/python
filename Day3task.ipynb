```python
#2 Movie Rating System 🎬
movies = [ ("Inception", "Sci-Fi", 8.8), ("The Godfather", "Drama", 9.2), ("Interstellar", "Sci-Fi", 8.6), ("Coco", "Animation",8.4), ("Parasite", "Drama", 8.6), ("Toy Story", "Animation",8.3), ]
a=[]
b=[]
c=[]
l=[]
d=set()
e={}
highest = {}
avg=0
for i in movies:
  if i[2]>=8.5:
    l.append(i[0])
    d.add(i[1])
  avg=avg+i[2]
print(l)
for j in movies:
  if j[1]=='Sci-Fi':
    a.append(j[0])
    e.update({j[1]:a})
  elif j[1]=='Drama':
    b.append(j[0])
    e.update({j[1]:b})
  else:
    c.append(j[0])
    e.update({j[1]:c})
print(e)
for n, g, r in movies:
    if g not in highest or r > highest[g][1]:
        highest[g] = (n, r)
print(highest)
print(avg/len(movies))
```

```python
#Q3 Course Enrollment Checker 🎓

python_class = {"Rahul", "Priya", "Arjun", "Sneha", "Karan"}
java_class = {"Priya", "Arjun", "Vikram", "Neha"}
ml_class = {"Sneha", "Vikram", "Rahul", "Divya"}
print(python_class.intersection(java_class))
a=python_class.difference(java_class)
print(a.difference(ml_class))
a=python_class.union(java_class)
print(a.union(ml_class))
a=python_class.intersection(java_class)
print(a.intersection(ml_class))
a=python_class.symmetric_difference(java_class)
print(a.symmetric_difference(ml_class))
```

```python
#5 Word Frequency Counter
fname='text.txt'
f=open(fname,'r')
data=f.read()
f.close()
data=data.split(' ')
a={}
b=[]
c=[]
d=[]
n=int(input('enter no of words'))
for i in data:
    if data.count(i)>=1:
        a.update({i:data.count(i)})
    if data.count(i)>1 and len(b)<n:
        b.append(i)
        d.append(data.count(i))
    e=list(zip(b,d))
    if data.count(i)==1:
        c.append(i)

print('Count_words')
print(a)
print('Top',n,'words')
print(e)
print('Words Appearing Only once')
print(c)
```
