```python
#Group movies by genre — return a dict: {"Sci-Fi": [...], "Drama": [...]}
movies = [ ("Inception", "Sci-Fi", 8.8), ("The Godfather", "Drama", 9.2), ("Interstellar", "Sci-Fi", 8.6), ("Coco", "Animation",8.4), ("Parasite", "Drama", 8.6), ("Toy Story", "Animation",8.3), ]
a=[]
b=[]
c=[]
d=set()
e={}
for i in movies:
  d.add(i[1])
for j in movies:
  if j[1]=='Sci-Fi':
    a.append(j[0])
    e.update({j[1]:a})
  elif j[1]=='Drama':
    b.append(j[0])
    e.update({j[1]:b})
  else:
    c.append(j[0])
    e.update({j[1]:c})
print(e)
```

```python
#Find the highest rated movie in each genre
movies = [ ("Inception", "Sci-Fi", 8.8), ("The Godfather", "Drama", 9.2), ("Interstellar", "Sci-Fi", 8.6), ("Coco", "Animation",8.4), ("Parasite", "Drama", 8.6), ("Toy Story", "Animation",8.3), ]
highest = {}

for n,g,r in movies:
    if g not in highest or r > highest[g][1]:
        highest[g] = (n, r)

for g, (n, r) in highest.items():
    print(g, " ", n, r)



```

```python
#Calculate the average rating of all movies
movies = [ ("Inception", "Sci-Fi", 8.8), ("The Godfather", "Drama", 9.2), ("Interstellar", "Sci-Fi", 8.6), ("Coco", "Animation",8.4), ("Parasite", "Drama", 8.6), ("Toy Story", "Animation",8.3), ]
avg=0
for i in movies:
  avg=avg+i[2]
print(avg/len(movies))
```
